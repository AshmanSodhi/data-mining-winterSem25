```python
import pandas as pd
import numpy as np
from datetime import datetime
print(datetime.now())
```

```python
df = pd.read_csv("synthetic_dataset.csv")
```

```python
# filling in missing values manually
df_enc = df.copy()
df_enc['Category'].fillna('Unknown',inplace = True)
df_enc['Price'].fillna(0,inplace = True)
df_enc['Rating'].fillna(0,inplace = True)
df_enc['Stock'].fillna('Unknown',inplace = True)
df_enc['Discount'].fillna(0,inplace = True)
print(f"Missing Values after : {df_enc.isnull().sum()}")
print(datetime.now())
```

```python
# find and replace encoding

encoding_map = {
    'A': 0,
    'B': 1,
    'C': 2,
    'D': 3,
    'Unknown' : 4
}

df_enc['Category_FR'] = df_enc['Category'].map(encoding_map)
print(f"Sample Encoded Data : {df_enc[['Category','Category_FR']].head()}")
print(datetime.now())
```

```python
# min max normalisation

df_norm = df_enc.copy()
min_val = df_enc['Price'].min()
max_val = df_enc['Price'].max()
df_norm['Price_MinMax'] = (df_norm['Price'] - min_val)/(max_val - min_val)
print(f"Original Range = {min_val} to {max_val}")
print(f"Normalised Range = {df_norm['Price_MinMax'].min()} to {df_norm['Price_MinMax'].max()}")
print(datetime.now())
```

```python
df_norm.to_csv('syn_Data_prepared.csv')
```
