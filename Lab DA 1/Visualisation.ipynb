```python
import pandas as pd
import numpy as np
from datetime import datetime
print(datetime.now())
```

```python
df = pd.read_csv("synthetic_dataset.csv")
print(datetime.now())
```

```python
#missing values fixing
df_d = df.copy()

#mean for rating, tried mean, but got negative values, hence going for median
df_d['Rating'].fillna(df_d['Rating'].median(),inplace = True)

#median for price and discount
df_d['Price'].fillna(df_d['Price'].median(),inplace = True)
df_d['Discount'].fillna(df_d['Discount'].median(),inplace = True)

#mode for categorical data
df_d['Stock'].fillna(df_d['Stock'].mode()[0], inplace=True)
df_d['Category'].fillna(df_d['Category'].mode()[0], inplace=True)
print(datetime.now())
```

```python
#a. Box Plot
import matplotlib.pyplot as plt
df_d.boxplot(column = 'Price', by='Category', ax=plt.gca())
plt.title('Price by Category')
plt.xlabel('Column')
plt.ylabel('Price')
```

```python
#b. Histogram
plt.hist(df_d['Price'].dropna(), bins=30, color='red', edgecolor='black')
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
print(datetime.now())
```

```python
#c. Scatter Plot
plt.scatter(df['Price'], df['Rating'], alpha=0.5, c='red')
plt.title('Price vs Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
print(datetime.now())
```

```python
#d. PieChart
category_counts = df_d['Category'].value_counts()
plt.pie(category_counts, labels = category_counts.index, autopct='%1.1f%%')
plt.title('Category Distribution')
print(datetime.now())
```
