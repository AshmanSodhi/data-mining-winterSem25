```python
import pandas as pd
import numpy as np
from datetime import datetime
print(datetime.now())
```

```python
df = pd.read_csv("synthetic_dataset.csv")
print(datetime.now())
```

```python
#b. fill in manually
df_enc = df.copy()
df_enc['Category'].fillna('Unknown',inplace = True)
df_enc['Price'].fillna(0,inplace = True)
df_enc['Rating'].fillna(0,inplace = True)
df_enc['Stock'].fillna('Unknown',inplace = True)
df_enc['Discount'].fillna(0,inplace = True)
print(f"Missing Values after : {df_enc.isnull().sum()}")
print(datetime.now())
```

```python
#a. Find and Replace Encoding

encoding_map = {
    'A': 0,
    'B': 1,
    'C': 2,
    'D': 3,
    'Unknown' : 4
}

df_enc['Category_FR'] = df_enc['Category'].map(encoding_map)
print(f"Sample Encoded Data : {df_enc[['Category','Category_FR']].head()}")
print(datetime.now())
```

```python
#b. Label Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_enc['Category_Label'] = le.fit_transform(df_enc['Category'])
print(f"Classes : {le.classes_}")
print(f"Sample Values : Sample Encoded Data : {df_enc[['Category','Category_Label']].head()} ")
print(datetime.now())
```

```python
#c. One Hot Encoding using Pandas

df_pandas_encoded = pd.get_dummies(df_enc, columns=['Category'], drop_first=True)
print(f"One-Hot Encoded Data using Pandas:\n{df_pandas_encoded.head()}\n")
print(datetime.now())
```

```python
#c. One Hot Encoding using ScikitLearn
from sklearn.preprocessing import OneHotEncoder
categorical_columns = df_enc.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[categorical_columns])

one_hot_df = pd.DataFrame(one_hot_encoded, 
                          columns=encoder.get_feature_names_out(categorical_columns))

df_sklearn_encoded = pd.concat([df.drop(categorical_columns, axis=1), one_hot_df], axis=1)

print(f"One-Hot Encoded Data using Scikit-Learn:\n{df_sklearn_encoded.head()}\n")

print(datetime.now())
```

```python
#d. Binary Encoding
encoding_map = {
    'A': 0,
    'B': 1,
    'C': 2,
    'D': 3,
    'Unknown' : 4
}

df_enc['Category_Bin'] = df_enc['Category'].map(encoding_map)
df_enc['Category_Bin'] = df_enc['Category_Bin'].apply(lambda x: format(x, 'b'))

print(f"Data after Binary Encoding : {df_enc[['Category','Category_Bin']].head()}")
print(datetime.now())
```
