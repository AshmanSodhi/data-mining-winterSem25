```python
#a. ignore the tupple - trying on column car

df_a = df.dropna()

print(f"Original Count: {len(df)}")
print(f"Row Count after dropping mising value tuple: {len(df_a)}")
print(datetime.now())
```

```python
#b. fill in manually
df_b = df.copy()
df_b['Category'].fillna('Unknown',inplace = True)
df_b['Price'].fillna(0,inplace = True)
df_b['Rating'].fillna(0,inplace = True)
df_b['Stock'].fillna('Unknown',inplace = True)
df_b['Discount'].fillna(0,inplace = True)
print(f"Missing Values after : {df_b.isnull().sum()}")
print(datetime.now())
```

```python
df_b
```

```python
#c. global constnat 
df_c = df.copy()
glo_text_con = "MISSING"
glo_con = -999
df_c['Category'].fillna(glo_text_con,inplace = True)
df_c['Price'].fillna(glo_con,inplace = True)
df_c['Rating'].fillna(glo_con,inplace = True)
df_c['Stock'].fillna(glo_text_con,inplace = True)
df_c['Discount'].fillna(glo_con,inplace = True)
print(f"Missing Values after : {df_c.isnull().sum()}")
print(datetime.now())
```

```python
#d. Central Tendency
df_d = df.copy()

#mean for rating, tried mean, but got negative values, hence going for median
df_d['Rating'].fillna(df_c['Rating'].median(),inplace = True)

#median for price and discount
df_d['Price'].fillna(df_d['Price'].median(),inplace = True)
df_d['Discount'].fillna(df_d['Discount'].median(),inplace = True)

#mode for categorical data
df_d['Stock'].fillna(df_d['Stock'].mode()[0], inplace=True)
df_d['Category'].fillna(df_d['Category'].mode()[0], inplace=True)
print(datetime.now())
print(df_d.head())
```
