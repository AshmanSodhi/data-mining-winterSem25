```python
import pandas as pd
import numpy as np
from datetime import datetime
print(datetime.now())
```

```python
df = pd.read_csv("synthetic_dataset.csv")
print(datetime.now())
```

```python
#MinMax Normalisation
df_norm = df.copy()
min_val = df['Price'].min()
max_val = df['Price'].max()
df_norm['Price_MinMax'] = (df_norm['Price'] - min_val)/(max_val - min_val)
print(f"Original Range = {min_val} to {max_val}")
print(f"Normalised Range = {df_norm['Price_MinMax'].min()} to {df_norm['Price_MinMax'].max()}")
print(datetime.now())
```

```python
#ZScore Normalisaton
df_norm = df.copy()
mean_val = df_norm['Price'].mean()
std_Val = df_norm['Price'].std()
df_norm['Price_ZScore'] = (df_norm['Price']-mean_val) / std_Val
print(f"Z Score Normalisation : mean = {mean_val} ||| Std = {std_Val}")
print(f"Normalised Mean : {df_norm['Price_ZScore'].mean()}")
print(f"Normalised Std : {df_norm['Price_ZScore'].std()}")
print(datetime.now())
```

```python
#c. Decimal Scaling
df_dec = df.copy()
max_abs = np.abs(df_dec['Price']).max()
j = len(str(int(max_abs)))
df_dec['Price_DS'] = df_dec['Price'] / (10 ** j)
print(f"Max Abs Value : {max_abs}")
print(f"Scaling Factor : 10^{j}")
print(f"Normalised Range : {df_dec['Price_DS'].min()} to {df_dec['Price_DS'].max()}")
print(df_dec['Price_DS'].head().values)
print(datetime.now())
```
