```python
import pandas as pd
import numpy as np
from datetime import datetime
print(datetime.now())
```

```python
df = pd.read_csv("synthetic_dataset.csv")
print(datetime.now())
```

```python
df_bin = df.copy()
n_bins = 5
```

```python
#a - Equal Frequency Binning

df_bin['Price_EqualFreq'] = pd.qcut(df_bin['Price'], q = n_bins, labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High'], duplicates='drop')
print(df_bin['Price_EqualFreq'].value_counts().sort_index())
print(datetime.now())
```

```python
#a - Equal Width Binning

df_bin['Price_EqualWidth'] = pd.cut(df_bin['Price'], bins = n_bins, labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High'], duplicates='drop')
print(df_bin['Price_EqualWidth'].value_counts().sort_index())
print(datetime.now())
```
