```python
# Q1 - Chi Square Test
from scipy.stats import chi2_contingency

cont_table = pd.crosstab(df['Category'], df['Stock'])
print(cont_table)

c2, p, dof, expected = chi2_contingency(cont_table)

print("Chi-square Statistic:", c2)
print("p-value:", p)
print("Degrees of Freedom:", dof)

if p < 0.05:
    print("Reject Null Hypothesis: Survival depends on Gender")
else:
    print("Fail to Reject Null Hypothesis: Category is independent of Stock")
    

print("\n")
print(datetime.now())
```

```python
# Q2 -Correlation (between Age and Fare - Numeric Columns)
num_df = df[['Price', 'Rating']].dropna()
```

```python
#a. Pearson
pear_corr = num_df.corr(method='pearson')
print("Pearson Correlation:\n", pear_corr)
```

```python
#b. Kendall Rank
kend_rank = num_df.corr(method='kendall')
print("Kendall Correlation:\n", kend_rank)
```

```python
#c. Spearman 
sm_corr = num_df.corr(method='spearman')
print("Spearman Correlation:\n", sm_corr)
print("\n")
print(datetime.now())
```

```python
# Q3 - Covariance
covmat = num_df.cov()
print("Covariance Matrix:\n", covmat)
print("\n")
print(datetime.now())
```

```python
# Q4 - Similarity and Dissimilarity
from scipy.spatial.distance import euclidean, cityblock, minkowski
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(num_df)
```

```python
# Calculating Distance for first two rows
euclidean_dist = euclidean(scaled_data[0], scaled_data[1])
manhattan_dist = cityblock(scaled_data[0], scaled_data[1])
minkowski_dist = minkowski(scaled_data[0], scaled_data[1], p=3)

print(f"Euclidean: {euclidean_dist}, Manhattan: {manhattan_dist}, Minkowski: {minkowski_dist}")
print("\n")
print(datetime.now())
```
